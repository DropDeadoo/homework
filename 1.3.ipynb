2. Определите количество четных элементов в последовательности, завершающейся числом 0

```python
x = int(input())
o = 0
n = 0
while x > 0:
  if x % 2 == 0:
    o += 1
  else:
    n += 1
  x = x // 10
print('Четные -', o)
print('Нечетные -', n)
```

3. Последовательность чисел вводится до 0. Найти мин и макс.
№исправлено

```python
x = int(input())

max = 0
min = 0

while x != 0:
    if x > max:
        max = x
    else:
        min = x
    x = int(input())
print("Максимум", max)
print("Минимум",min)
```

4. Последовательность состоит из различных натуральных чисел и завершается 0. Опр. значение 2-го по величине элемента в этой последовательности. В последовательности должно быть хотя бы 2

```python
a = int(input())
max1 = 0
max2 = 0

while a != 0:
    if a > max1:
        max2 = max1
        max1 = a
    elif a > max2:
        max2 = a
    a = int(input())
print (max2)
```

5. N факториал какого числа?
#исправлено

```python
n = int(input())
f = 1 #факториал
x = 1
while (x <= n):
    f = f * x
    x += 1
print("Факториал числа", n,"равен", f)
```

6. Дана последовательность натуральных чисел, заверш. числом 0. Определите, какое наибольшее число подряд идущих элементов этой последовательности равны друг другу.
#исправлено

```python
a0 = -1
l = 0 #текущая длина
n = 0 #искомая длина
a = int(input()) 
while a != 0:
    if a == a0:
        l += 1
    else:
        a0 = a
        n = max(n,l)
        l = 1
    a = int(input()) 
n = max(n,l)
print('Ответ: ',n)
```

b. По данному натуральному числу N найдите наибольшую целую степнь двойки, не превосходящую N. Выведите показатель степени и саму степень. Операцией возведения в степень пользоваться нельзя!
#исправлено

```python
n = int(input()) 
b = 2 #степень
i = 1 #после возведения

while b <= n:
  b *= 2
  i += 1
print(i - 1, b // 2)
```
