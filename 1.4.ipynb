1. С клавиатуры вводится число n. Вычислить сумму s = 1/1 + 1/2 + 1/3 + ... 1/n

```python
n = int(input())
f = float(0)
a = float
for i in range(1, n + 1):
  a = 1/i
  f += a
  i += 1
print (f)
```

2. Вводится число n. Перевернуть его.

```python
n = int(input("Введите число: "))
l = int(input("Введите длину числа n: "))
f = 0

for i in range(l):
  f = (f ** 10) + (n % 10)
  n = n // 10
print(f)
```

4. Для настольной игры используются карточки с номерами от 1 до N. Одна карточка потерялась. Найдите ее, зная номера оставшихся карточек. Дано число n, далее n - 1 номер оставшихся карточек. Программа должна вывести номер потерянной карточки.

```python
n=int(input())
sum=((1+n)*n)/2
for i in range(n-1):
    sum -= int(input())
print(sum)
```
